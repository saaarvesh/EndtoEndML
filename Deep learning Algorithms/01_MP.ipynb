### Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('class', axis=1)
Y = data['class']
```

```python
#by stratifying data, data will be split in such a way that ratio of 0 and 1 class will remain almost same in test and train, same for 
#other values, increases representativeness of each class
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify=Y)
```

```python
X.shape, X_train.shape, X_test.shape
     
```

```python
# because of statify sampling mean of train and test will be very close to each other for each feature
Y.mean(), Y_train.mean(), Y_test.mean()
```
